```python
import pandas as pd
import numpy as np
#lectura de datos por categoria
dataset=pd.read_csv("WineQT.csv")
dataset.head(1150)
```

```python
# Extraer algunas columnas del dataset WineQT para aplicar las penalizaciones L1 y L2
columns_for_weights = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides']
weights = dataset[columns_for_weights].mean().tolist()  # Tomar la media de cada columna como ejemplo de coeficientes

# Función para normalizar los datos (min-max normalization)
def normalize(data):
    min_val = min(data)
    max_val = max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]

# Función para calcular la penalización L1
def l1_penalty(weights, lambda_value):
    return lambda_value * sum(abs(w) for w in weights)

# Función para calcular la penalización L2
def l2_penalty(weights, lambda_value):
    return lambda_value * sum(w**2 for w in weights)

# Normalizar los coeficientes del dataset WineQT
weights_normalized = normalize(weights)

# Parámetro de regularización (lambda)
lambda_value = 0.1

# Calcular penalización L1
l1 = l1_penalty(weights_normalized, lambda_value)

# Calcular penalización L2
l2 = l2_penalty(weights_normalized, lambda_value)

l1, l2  # Mostrar resultados de las penalizaciones L1 y L2
```
