```python
import pandas as pd
import numpy as np
#lectura de datos por categoria
dataset=pd.read_csv("WineQT.csv")
dataset.head(1150)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Excluir la columna 'id'
columnas_seleccionadas = dataset.drop(columns=['Id'])  # Excluir 'id'

# Calcular media, mediana y moda para cada columna
media = columnas_seleccionadas.mean()
mediana = columnas_seleccionadas.median()
moda = columnas_seleccionadas.mode().iloc[0]  # Tomar la primera moda si hay más de una

# Mostrar los resultados
formatted_output = (
    f"Media:\n{media.to_string(index=True, float_format='%.4f')}\n\n"
    f"Mediana:\n{mediana.to_string(index=True, float_format='%.4f')}\n\n"
    f"Moda:\n{moda.to_string(index=True, float_format='%.4f')}\n"
)
print(formatted_output)

# Graficar diagrama de cajas-bigote
plt.figure(figsize=(12, 6))
sns.boxplot(data=columnas_seleccionadas.iloc[:, :12])
plt.title('Diagrama de Cajas-Bigote')
plt.ylabel('Valores')
plt.xlabel('Características')
plt.ylim(0, 180)
plt.xticks(rotation=45)
plt.show()
```

**Explicación de Resultados:**

*Media:*
*   La media es el promedio de los valores de cada columna. Representa un punto de equilibrio, pero puede estar influenciada por valores extremos (outliers).
*   En tu dataset, por ejemplo, la media del alcohol es 10.44, lo que indica que la concentración promedio de alcohol en los vinos analizados es cercana a ese valor.

*Mediana:*
*   La mediana es el valor central cuando los datos están ordenados. Es menos sensible a valores atípicos.
*   Por ejemplo, la mediana del pH es 3.31, lo que indica que la mitad de los vinos tienen un pH menor o igual a este valor, y la otra mitad un pH mayor.

*Moda:*
*   La moda es el valor que más veces se repite en los datos. Es útil para identificar concentraciones comunes.
*   En el caso de citric acid, la moda es 0.00, lo que sugiere que muchos vinos no contienen ácido cítrico.

**Explicación del Diagrama de Cajas-Bigote:**
El gráfico de cajas-bigote proporciona una visualización compacta de la distribución de los datos de varias columnas y permite identificar rápidamente valores atípicos.

*Cajas:*
*   Representan el rango intercuartil (IQR), es decir, el rango entre el primer cuartil (Q1) y el tercer cuartil (Q3). Este rango contiene el 50% de los datos.
*   Las líneas dentro de las cajas representan la mediana de cada variable.

*Bigotes:*
Los bigotes muestran la extensión de los datos, hasta un máximo de 1.5 veces el IQR. Más allá de esto, se consideran valores atípicos (outliers), que son representados por puntos o círculos.

*Valores atípicos:*
Puedes observar que hay varios outliers en las columnas como free sulfur dioxide y total sulfur dioxide, lo que indica que algunos vinos tienen concentraciones de estos compuestos mucho más altas que la mayoría de los vinos.

**Ejemplos de Interpretación:**

*Fixed Acidity:*
*   Mediana: 7.90 (lo que significa que la mitad de los vinos tienen una acidez fija por debajo de este valor).
*   Outliers: No se observan muchos valores atípicos, lo que sugiere que la mayoría de los vinos tienen una acidez dentro de un rango razonable.

*Free Sulfur Dioxide:*
*   Mediana: 13.00.
*   Outliers: Hay muchos outliers que se extienden por encima de 30, lo que indica que algunos vinos tienen niveles inusualmente altos de dióxido de azufre libre.

*Alcohol:*
*   Mediana: 10.20.
*   Outliers: Los datos muestran algunos valores atípicos en el extremo superior, lo que significa que algunos vinos tienen niveles de alcohol significativamente más altos que la mayoría.

**Conclusión:**
*   Distribución: El gráfico y las medidas indican que varias columnas, como el dióxido de azufre libre y total, contienen valores atípicos que distorsionan la media, lo que explica la diferencia entre la media y la mediana en estas variables.
*   Dispersión: El rango intercuartil (IQR) muestra que algunas variables están mucho más dispersas que otras, como el dióxido de azufre total comparado con la acidez volátil.
*   Relevancia de los outliers: Los valores atípicos en las columnas relacionadas con el dióxido de azufre y el alcohol son de especial interés, ya que estos compuestos influyen significativamente en la calidad del vino y podrían representar productos fuera de los estándares habituales.
