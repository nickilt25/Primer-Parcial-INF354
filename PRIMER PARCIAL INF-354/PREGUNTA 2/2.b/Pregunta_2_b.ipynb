```python
import pandas as pd
import numpy as np
#lectura de datos por categoria
dataset=pd.read_csv("WineQT.csv")
dataset.head(1150)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionar tres columnas relevantes
columnas_seleccionadas = dataset[['alcohol', 'pH', 'quality']]  # Asegúrate de que las columnas existan en el dataset

# Descripción de las columnas seleccionadas
print(columnas_seleccionadas.describe())

# Gráfico de dispersión entre 'alcohol' y 'quality'
plt.figure(figsize=(8,6))
plt.scatter(dataset['alcohol'], dataset['quality'], color='blue', alpha=0.5)
plt.title('Relación entre Alcohol y Calidad')
plt.xlabel('Alcohol')
plt.ylabel('Calidad')
plt.grid(True)
plt.show()

# Gráfico de dispersión entre 'pH' y 'quality'
plt.figure(figsize=(8,6))
plt.scatter(dataset['pH'], dataset['quality'], color='green', alpha=0.5)
plt.title('Relación entre pH y Calidad')
plt.xlabel('pH')
plt.ylabel('Calidad')
plt.grid(True)
plt.show()

# Mapa de calor de correlaciones
plt.figure(figsize=(10, 8))
sns.heatmap(columnas_seleccionadas.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor de Correlaciones entre Alcohol, pH y Calidad')
plt.show()
```

1. **Alcohol:**

*   *Rango:* Los valores de alcohol tienden a estar en un rango relativamente alto(por ejemplo, entre 8% y 14% en vinos).
*   *Relación con la calidad:* A menudo, un mayor contenido de alcohol puede estar asociado con una mayor calidad percibida, aunque no siempre es así.
*   *Dispersión:* La distribución de alcohol puede mostrar una leve variabilidad.
*   *Relación con otras variables:* Podría correlacionarse con otras características como el azúcar residual o la densidad.

2. **pH:**

*   *Acidez:* Un pH más bajo indica un vino más ácido. El pH promedio suele estar entre 3.0 y 3.5 para la mayoría de los vinos.
*   *Relación con la calidad:* La acidez afecta el sabor del vino, y por lo tanto, la percepción de su calidad.
*   *Variabilidad:* El pH puede mostrar variabilidad en relación con otras características, como la acidez volátil.
*   *Distribución:* El pH podría tener una distribución normal o sesgada dependiendo del tipo de vino.

3. **Calidad:**

*   *Categórica:* Aunque es una variable numérica, la calidad puede verse como una variable categórica (escalas de 0 a 10).
*   *Relación con otras variables:* La calidad puede correlacionarse positivamente con el contenido de alcohol y negativamente con el pH.
*   *Escala discreta:* Normalmente es evaluada en una escala discreta (por ejemplo, de 0 a 10).
*   *Relevancia para el análisis:* La calidad es probablemente la variable dependiente en muchos análisis, ya que se busca predecir o explicar qué características del vino influyen en su calidad.
